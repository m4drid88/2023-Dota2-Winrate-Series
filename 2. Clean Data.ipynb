```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json("2023riyad-dota2series.json").rename(columns={"id":"serieId"})
df.head()
```

```python
df.info()
```

```python
df = (df.assign(**df.teamOne.apply(pd.Series))).rename(columns={"id":"teamOneId","name":"teamOneName"}).drop("teamOne",axis=1)
df = (df.assign(**df.teamTwo.apply(pd.Series))).rename(columns={"id":"teamTwoId","name":"teamTwoName"}).drop("teamTwo",axis=1)
df["winningTeamId"] = df["winningTeamId"].fillna(0).astype(np.int64)
df["lastMatchDateTime"] = pd.to_datetime(df["lastMatchDateTime"],unit="s")
```

```python
df.info()
```

```python
df.sample(4)
```

## Reemplazar columnas

```python
leagues = {15438:"The Bali Major",15251:"The Berlin Major",15089:"The Lima Major",15196:"DreamLeague Season 19",15439:"DreamLeague Season 20"}
df["leagueId"] = df["leagueId"].replace(leagues)
df = df.rename(columns={"leagueId":"leagueName"})
```
