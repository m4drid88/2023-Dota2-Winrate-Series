```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("dataseries.xlsx",sheet_name="main")
df["type"] = df["type"].replace({"BEST_OF_FIVE":"bo5","BEST_OF_TWO":"bo2","BEST_OF_THREE":"bo3","BEST_OF_ONE":"bo1"})
df.head()
```

```python
dfTeam1 = df.copy().rename(columns={"teamOneId":"teamId","teamOneName":"teamName","teamTwoId":"enemyId","teamTwoName":"enemyName","teamOneWinCount":"teamWinCount","teamTwoWinCount":"enemyWinCount"})
dfTeam2 = df.copy().rename(columns={"teamTwoId":"teamId","teamTwoName":"teamName","teamOneId":"enemyId","teamOneName":"enemyName","teamOneWinCount":"enemyWinCount","teamTwoWinCount":"teamWinCount"})

df2 = pd.concat([dfTeam1,dfTeam2],ignore_index=True)
df2
```

```python
df2["serieResult"] = df2.apply(lambda x : "draw" if x["winningTeamId"] == 0 else ("win" if x["winningTeamId"] == x["teamId"] else "lose"), axis= 1)
df2
```

```python
df2["teamName"] = df2["teamName"].replace({"BB Team":"BetBoom Team"})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
```

```python
teams = df2["teamName"].unique()
teams
```

```python
def generateChart(team):
    # team = "Gaimin Gladiators"
    leagues = ['The Lima Major',
            #    'DreamLeague Season 19',
            'The Berlin Major',
            #    'DreamLeague Season 20,
            'The Bali Major',
            "Riyadh Masters 2023",
            ]
    plt.figure(figsize=(12,5))
    plt.suptitle(f"Resultado por series : {team}")

    for i,l in enumerate(leagues,start=1):
        plt.subplot(1,len(leagues),i)
        dfTeam = df2.query("teamName == @team & leagueName == @l")
        sns.countplot(dfTeam,hue="serieResult",x="type",order=["bo1","bo2","bo3","bo5"],hue_order=["win","draw","lose"],palette=("#27AE60","#3498DB","#E74C3C"))
        plt.title(l)
        plt.yticks(np.arange(0,11,1))    
        plt.legend(title="Resultado",loc="upper right")
        plt.ylabel("Series")
        plt.xlabel("")

    plt.tight_layout()
    plt.savefig(f"series_{team}.pdf",format="pdf",dpi=120)
```

```python
for team in teams:
    generateChart(team)
```
